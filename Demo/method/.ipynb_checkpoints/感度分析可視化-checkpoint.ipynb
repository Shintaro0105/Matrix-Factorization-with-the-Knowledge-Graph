```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('dim128-30-30neg1_1.csv')
df2 = pd.read_csv('dim128-30-30neg1_2.csv')
df3 = pd.read_csv('dim128-30-30neg1_3.csv')
```

```python
x=df1['reg_v']
p1=df1['pre10']
r1=df1['rec10']
p2=df2['pre10']
r2=df2['rec10']
p3=df3['pre10']
r3=df3['rec10']
```

```python
plt.xlabel("Beta")
plt.ylabel("Precision@10")
plt.plot(x,p1,marker="o",label="Alpha=0.0")
plt.plot(x,p2,marker="o",label="Alpha=0.1")
plt.plot(x,p3,marker="o",label="Alpha=1.0")
plt.legend()
```

```python
plt.xlabel("Beta")
plt.ylabel("Recall@10")
plt.plot(x,r1,marker="o",label="Alpha=0.0")
plt.plot(x,r2,marker="o",label="Alpha=0.1")
plt.plot(x,r3,marker="o",label="Alpha=1.0")
plt.legend()
```
